# Turbo Row Example

```python
import numpy as np
from parafoil import CamberThicknessAirfoil, TurboRowPassage

airfoil = CamberThicknessAirfoil(
    inlet_angle=np.radians(40.0),
    outlet_angle=np.radians(-20.0),
    stagger_angle=np.radians(0.0),
    upper_thick_prop=[0.035, 0.14, 0.14, 0.035, 0.007],
    lower_thick_prop=[0.035, 0.14, 0.14, 0.035, 0.007],
    leading_prop=0.5,
    trailing_prop=0.5,
    chord_length=1
)

passage = TurboRowPassage(
    airfoil=airfoil,
    spacing_to_chord=1.0,
    leading_edge_gap_to_chord=0.25,
    trailing_edge_gap_to_chord=0.25,
)

airfoil.visualize(include_camber_ctrl_pnts=True)
passage.visualize()
```

## Rotor Airfoil

```python
import numpy as np
from parafoil import CamberThicknessAirfoil, TurboRowPassage


airfoil = CamberThicknessAirfoil(
    inlet_angle=np.radians(-55),
    outlet_angle=np.radians(25),
    stagger_angle=np.radians(-20),
    upper_thick_prop=[0.025, 0.05, 0.1, 0.05, 0.025],
    lower_thick_prop=[0.025, 0.05, 0.1, 0.05, 0.025],
    leading_prop=0.5,
    trailing_prop=0.5,
    chord_length=1
)

airfoil.visualize(include_camber_ctrl_pnts=True)

passage = TurboRowPassage(
    airfoil=airfoil,
    spacing_to_chord=1.0,
    leading_edge_gap_to_chord=0.25,
    trailing_edge_gap_to_chord=0.25,
    num_airfoils=2,
    type="surface"
)

passage.visualize()
```
