# Airfoil Example

```python
import numpy as np
from parafoil.airfoil import Airfoil
from parafoil.passage import PassageBuilder

airfoil = Airfoil(
    inlet_angle=np.radians(40.0),
    outlet_angle=np.radians(20.0),
    stagger_angle=np.radians(0.0),
    upper_thick_dist=[0.05, 0.2, 0.2, 0.05, 0.01],
    lower_thick_dist=[0.05, 0.2, 0.2, 0.05, 0.01],
    leading_prop=0.5,
    trailing_prop=0.5,
    chord_length=1.44
)

airfoil.visualize(include_camber_ctrl_pnts=True)

passage = PassageBuilder.airfoil_row_passage(
    airfoil=airfoil,
    spacing=1.0 * airfoil.chord_length,
    leading_edge_gap=0.25,
    trailing_edge_gap=0.25,
    num_blades=1,
)

passage.visualize()
```

## Rotor Airfoil

```python
import numpy as np
from parafoil.airfoil import Airfoil
from parafoil.passage import PassageBuilder

airfoil = Airfoil(
    inlet_angle=np.radians(-55),
    outlet_angle=np.radians(25),
    stagger_angle=np.radians(-20),
    upper_thick_dist=[0.025, 0.05, 0.1, 0.05, 0.025],
    lower_thick_dist=[0.025, 0.05, 0.1, 0.05, 0.025],
    leading_prop=0.5,
    trailing_prop=0.5,
    chord_length=1
)

airfoil.visualize(include_camber_ctrl_pnts=True)

passage = PassageBuilder.airfoil_row_passage(
    airfoil=airfoil,
    spacing=1.0 * airfoil.chord_length,
    leading_edge_gap=0.25,
    trailing_edge_gap=0.25,
    num_blades=2
)

passage.visualize()
```
